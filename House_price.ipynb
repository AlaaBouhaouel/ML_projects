```python
df.drop(columns=['Alley','MiscFeature'],inplace=True)
df.head()
```

```python
import numpy as np
df['FireplaceQu']=df['FireplaceQu'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
```

```python
df.head()
```

```python
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
df.drop(columns=['YearBuilt'],inplace=True)
```

```python
df['PropertyFrontageDensity']=df['LotFrontage']/df['LotArea']
df.drop(columns=['LotFrontage','LotArea'],inplace=True)
```

```python
df['OverallScore']=df['OverallQual']*df['OverallCond']+df['YearRemodAdd']**2
df.drop(columns=['OverallQual','OverallCond','YearRemodAdd'],inplace=True)
```

```python
#ExteriorScore = ExterQual_num * ExterCond_num
df['ExterQual']=df['ExterQual'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
df['ExterCond']=df['ExterCond'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
```

```python
df['ExterScore']=df['ExterQual']*df['ExterCond']
df.drop(columns=['ExterQual','ExterCond'],inplace=True)
```

```python
df['BasementFinishedRatio']=(df['BsmtFinSF1']+df['BsmtFinSF2'])/df['TotalBsmtSF']
df.drop(columns=['BsmtFinSF1','BsmtFinSF2','TotalBsmtSF'],inplace=True)
```

```python
df['BasementBathroomScore']=df['BsmtFullBath']+0.5*df['BsmtHalfBath']
df.drop(columns=['BsmtFullBath','BsmtHalfBath'],inplace=True)
```

```python
df['TotalBathrooms']=df['FullBath']+0.5*df['HalfBath']+df['BasementBathroomScore']
df.drop(columns=['FullBath','HalfBath','BasementBathroomScore'],inplace=True)
```

```python
df['RoomPerBedroomRatio']=df['TotRmsAbvGrd']/df['BedroomAbvGr']
df.drop(columns=['TotRmsAbvGrd','BedroomAbvGr'],inplace=True)
```

```python
df['HasGarage'] = (df['GarageArea'] > 0).astype(int)
```

```python
df['GarageType']=df['GarageType'].replace(['Attchd', 'Detchd','BuiltIn','CarPort','Basment',np.nan],
                        [5,3,4,2,1,0])
```

```python
df['GarageQual']=df['GarageQual'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
df['GarageCond']=df['GarageCond'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
df['GarageFinish']=df['GarageFinish'].replace(['Fin', 'RFn','Unf',np.nan],
                        [3,2,1,0])
```

```python
# Replace NaN values in 'GarageYrBlt' with the corresponding values from 'YrSold'
df['GarageYrBlt'].fillna(df['YrSold'], inplace=True)
```

```python
df['GarageYrBlt']
```

```python
# Ensure all columns used in the calculation are numeric
df['GarageArea'] = pd.to_numeric(df['GarageArea'], errors='coerce').fillna(0)
df['GarageCars'] = pd.to_numeric(df['GarageCars'], errors='coerce').fillna(0)
df['GarageQual'] = pd.to_numeric(df['GarageQual'], errors='coerce').fillna(0)
df['GarageCond'] = pd.to_numeric(df['GarageCond'], errors='coerce').fillna(0)
df['GarageFinish'] = pd.to_numeric(df['GarageFinish'], errors='coerce').fillna(0)
df['GarageType'] = pd.to_numeric(df['GarageType'], errors='coerce').fillna(0)

df['Garage0']=(df['GarageArea']*0.4)+(df['GarageCars']*0.2)+(df['GarageQual']*0.1)+(df['GarageCond']*0.1)+(df['GarageFinish']*0.1)+(df['GarageType']*0.05)

# Define the columns that were subject to pd.to_numeric and fillna
columns_to_check = ['GarageArea', 'GarageCars', 'GarageQual', 'GarageCond', 'GarageFinish', 'GarageType']

# Check for NaN values in these columns after conversion and filling
for col in columns_to_check:
    nan_count = df[col].isnull().sum()
    if nan_count > 0:
        print(f"Column '{col}' still has {nan_count} NaN values after conversion and filling. This should not happen if fillna(0) was successful.")

# You can also check the data types of these columns to confirm they are numeric
print("\nData types of the checked columns after conversion and filling:")
print(df[columns_to_check].dtypes)
```

```python
df['Garagescore']=df['HasGarage']*(df['Garage0']// (1 +(-df['GarageYrBlt']+df['YrSold'])))
```

```python
df=df.drop(columns=['GarageArea','GarageCars','GarageQual','GarageCond','GarageFinish','GarageType','GarageYrBlt','HasGarage','Garage0'])
```

```python
df['TotalPorchSF']=df['OpenPorchSF']+df['EnclosedPorch']+df['3SsnPorch']+df['ScreenPorch']
df.drop(columns=['OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch'],inplace=True)
```

```python

# Assuming your DataFrame is called df
df['HasPool'] = (df['PoolArea'] > 0).astype(int)
df['HasDeck'] = (df['WoodDeckSF'] > 0).astype(int)
df['HasFence'] = (df['Fence'].notna() & (df['Fence'] != 'None')).astype(int)
```

```python
df.drop(columns=['WoodDeckSF','Fence'],inplace=True)
```

```python
df['HasCentralAir'] = (df['CentralAir'] == 'Y').astype(int)
```

```python
df.drop(columns=['CentralAir'],inplace=True)
```

```python
utilities_map = {'AllPub': 2, 'NoSeWa': 1, 'ELO': 0, 'NoSewr': 0, 'None': 0}
paved_drive_map = {'Y': 2, 'P': 1, 'N': 0}
df['Utilities_num'] = df['Utilities'].map(utilities_map).fillna(0)
df['PavedDrive_num'] = df['PavedDrive'].map(paved_drive_map).fillna(0)

df['UtilityScore'] = df['Utilities_num'] + df['PavedDrive_num']
```

```python
df['UtilityScore'].nunique()
```

```python
df.drop(columns=['Utilities','PavedDrive','Utilities_num','PavedDrive_num'],inplace=True)
```

```python
df['KitchenQual']=df['KitchenQual'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
```

```python
df['kitchenScore']=df['KitchenQual']*df['KitchenAbvGr']
df.drop(columns=['KitchenQual','KitchenAbvGr'],inplace=True)
```

```python
df['PoolQC']=df['PoolQC'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
```

```python
df['PoolArea'].value_counts()
```

```python
df['poolScore']=df['HasPool']*df['PoolArea']+df['PoolQC']
df.drop(columns=['PoolArea','HasPool','PoolQC'],inplace=True)
```

```python
df['poolScore'].value_counts()
```

```python
df['FireplaceQu']=df['FireplaceQu'].replace(['Ex', 'Gd','TA','Fa',np.nan],
                        [8,6,4,2,0])
```

```python
df['Fireplacescore']=df['Fireplaces']*df['FireplaceQu']
df.drop(columns=['Fireplaces','FireplaceQu'],inplace=True)
```

```python
df['SaleDate'] = df['YrSold'] + (df['MoSold'] - 1) / 12
df.drop(columns=['YrSold','MoSold'],inplace=True)
```

```python
df['SaleTypeEncoded'] = df['SaleType'].replace({
    'WD': 5, 'New': 4, 'COD': 3, 'Con': 2, 'Oth': 1,'ConLD':2,'ConLI':2,'ConLw':2,'CWD':4
})

df['SaleConditionEncoded'] = df['SaleCondition'].replace({
    'Normal': 5, 'Abnorml': 4, 'Partial': 3, 'AdjLand': 2, 'Alloca': 2,'Family':1
})
```

```python
df['SaleConditionEncoded'].value_counts()

```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['SalePriceNormalized'] = scaler.fit_transform(df[['SalePrice']])
```

```python
df['SaleScore'] = (df['SaleTypeEncoded'] * 0.4) + (df['SaleConditionEncoded'] * 0.3) + (df['SalePriceNormalized'] * 0.3)
```

```python
df.drop(columns=['SaleTypeEncoded','SaleConditionEncoded','SalePriceNormalized','SaleType','SaleCondition'],inplace=True)
```

```python
df['BsmtQualEncoded'] = df['BsmtQual'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, np.nan: 0})
df['BsmtCondEncoded'] = df['BsmtCond'].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, np.nan: 0})
df['BsmtExposureEncoded'] = df['BsmtExposure'].replace({'Gd': 3, 'Av': 2, 'Mn': 1, 'No': 0, np.nan: 0})
df['BsmtFinType1Encoded'] = df['BsmtFinType1'].replace({'GLQ': 5, 'ALQ': 4, 'BLQ': 3, 'Rec': 2, 'LwQ': 1, 'Unf': 0, np.nan: 0})
df['BsmtFinType2Encoded'] = df['BsmtFinType2'].replace({'GLQ': 5, 'ALQ': 4, 'BLQ': 3, 'Rec': 2, 'LwQ': 1, 'Unf': 0, np.nan: 0})
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['BsmtUnfSFNormalized'] = scaler.fit_transform(df[['BsmtUnfSF']])
```

```python
df['BasementScore'] = (df['BsmtQualEncoded'] * 0.3) + (df['BsmtCondEncoded'] * 0.2) + \
                      (df['BsmtExposureEncoded'] * 0.2) + (df['BsmtFinType1Encoded'] * 0.2) + \
                      (df['BsmtFinType2Encoded'] * 0.1) + (df['BsmtUnfSFNormalized'] * 0.2)
```

```python
df.drop(columns=['BsmtQualEncoded','BsmtExposureEncoded','BsmtFinType2Encoded','BsmtUnfSFNormalized','BsmtFinType2','BsmtFinType1','BsmtExposure','BsmtCond','BsmtQual'],inplace=True)
```

```python
df['StreetEncoded'] = df['Street'].replace({'Pave': 1, 'Grvl': 0})
df['LotShapeEncoded'] = df['LotShape'].replace({'Reg': 3, 'IR1': 2, 'IR2': 1, 'IR3': 0})
df['LandContourEncoded'] = df['LandContour'].replace({'Lvl': 3, 'Bnk': 2, 'HLS': 1, 'Low': 0})
df['LotConfigEncoded'] = df['LotConfig'].replace({'Inside': 3, 'Corner': 2, 'CulDSac': 1, 'FR2': 1, 'FR3': 0})
df['LandSlopeEncoded'] = df['LandSlope'].replace({'Gtl': 2, 'Mod': 1, 'Sev': 0})
df['Condition1Encoded'] = df['Condition1'].replace({'Norm': 3, 'Artery': 2, 'Feedr': 1, 'RRNn': 0, 'RRAn': -1,'RRNe': -2,'RRAe': -2,  'PosN': 4, 'PosA': 5})
df['Condition2Encoded'] = df['Condition2'].replace({'Norm': 3, 'Artery': 2, 'Feedr': 1, 'RRNn': 0, 'RRAn': -1, 'RRNe': -2,'RRAe': -2, 'PosN': 4, 'PosA': 5})
```

```python
df['Condition2Encoded'].value_counts()
```

```python
df['NeighborhoodEncoded'] = df.groupby('Neighborhood')['SalePrice'].transform('mean')
```

```python
df['PropertyAccessibilityScore'] = (df['StreetEncoded'] * 0.2) +  \
                                   (df['LotShapeEncoded'] * 0.1) + (df['LandContourEncoded'] * 0.1) + \
                                   (df['LotConfigEncoded'] * 0.1) + (df['LandSlopeEncoded'] * 0.1) + \
                                   (df['Condition1Encoded'] * 0.1) + (df['Condition2Encoded'] * 0.1) + \
                                   (df['NeighborhoodEncoded'] * 0.3)
```

```python
df.drop(columns=['StreetEncoded','LotShapeEncoded','LandContourEncoded','LotConfigEncoded','LandSlopeEncoded','Condition1Encoded','Condition2Encoded','NeighborhoodEncoded'],inplace=True)
```

```python
df.drop(columns=['Id','Neighborhood','Condition1','Condition2'],inplace=True)
```

```python
df.drop(columns=['BsmtUnfSF'],inplace=True)
```

```python
df['BldgType']=df['BldgType'].replace({'1Fam': 5, 'TwnhsE': 4, 'Twnhs': 3, 'Duplex': 2, '2fmCon': 1})
df['HouseStyle']=df['HouseStyle'].replace({'2Story': 5, '1Story': 4, '1.5Fin': 3, 'SLvl': 2, 'SFoyer': 1,'1.5Unf': 2,'2.5Unf': 2,'2.5Fin': 3})
df['Foundation']=df['Foundation'].replace({'PConc': 5, 'Wood': 4,'CBlock': 4, 'BrkTil': 3, 'Slab': 2, 'Stone': 1})
```

```python
df['BuildingScore'] = df['BldgType'] + df['HouseStyle']+df['Foundation']
```

```python
df.drop(columns=['BldgType','HouseStyle','Foundation'],inplace=True)
```

```python
df['ExteriorScore'] = df['RoofStyle'].replace({'Hip': 5, 'Gable': 4, 'Flat': 3,'Gambrel': 3, 'Mansard': 2, 'Shed': 1}) + \
                      df['RoofMatl'].replace({'CompShg': 5, 'WdShngl': 4,'WoodShg': 4, 'ClyTile': 3,'Membran': 2,'Tar&Grv': 2, 'WdShake': 2, 'Metal': 1,'Roll': 1}) + \
                      df['Exterior1st'].replace({'VinylSd': 5,'HdBoard': 4,'MetalSd': 4,'Wd Sdng': 3,'Plywood': 3,'CemntBd': 4,'BrkFace': 5,'WdShing': 3,'Stucco': 3, 'AsbShng': 2,'BrkComm': 5,'Stone': 5, 'AsphShn': 1,'ImStucc': 2,'CBlock': 2 })
```

```python
df.drop(columns=['RoofStyle','RoofMatl','Exterior1st'],inplace=True)
```

```python
df['UtilityScore'] = df['Heating'].replace({'GasA': 5, 'GasW': 4, 'Grav': 3, 'Wall': 2, 'OthW': 1, 'Floor': 0}) + df['HeatingQC'].replace({'Ex': 5, 'TA': 4, 'Gd': 3, 'Fa': 2, 'Po': 1}) + df['Electrical'].replace({'SBrkr': 5, 'FuseA': 4, 'FuseF': 3, 'FuseP': 2, 'Mix': 1})
```

```python
df.drop(columns=['Heating','HeatingQC','Electrical'],inplace=True)
```

```python
df['LivingSpaceScore'] = df['1stFlrSF'] + df['2ndFlrSF'] + df['GrLivArea'] - df['LowQualFinSF']
```

```python
df.drop(columns=['1stFlrSF','2ndFlrSF','GrLivArea','LowQualFinSF'],inplace=True)
```

```python
df['MasVnrType'].value_counts()
```

```python
df['MasonryScore'] = df['MasVnrType'].replace({'BrkFace': 5, 'Stone': 4,'BrkCmn': 3, 'None': 0}) + df['MasVnrArea'].fillna(0)
df['MiscScore'] = df['MiscVal'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df.drop(columns=['MasVnrType','MasVnrArea','MiscVal'],inplace=True)
```

```python
df['MSZoningEncoded'] = df['MSZoning'].replace({'FV': 5, 'RL': 4, 'RH': 3, 'RM': 2, 'C (all)': 1})
df['StreetEncoded'] = df['Street'].replace({'Pave': 1, 'Grvl': 0})
df['LotShapeEncoded'] = df['LotShape'].replace({'Reg': 3, 'IR1': 2, 'IR2': 1, 'IR3': 0})
df['LandContourEncoded'] = df['LandContour'].replace({'Lvl': 3, 'Bnk': 2, 'HLS': 1, 'Low': 0})
df['LotConfigEncoded'] = df['LotConfig'].replace({'Inside': 3, 'Corner': 2, 'CulDSac': 1, 'FR2': 1, 'FR3': 0})
df['LandSlopeEncoded'] = df['LandSlope'].replace({'Gtl': 2, 'Mod': 1, 'Sev': 0})
df['Exterior2ndEncoded'] = df['Exterior2nd'].replace({'VinylSd': 5, 'HdBoard': 4, 'MetalSd': 4,'Wd Shng': 3,  'Wd Sdng': 3, 'Plywood': 3, 'CmentBd': 4, 'BrkFace': 5, 'WdShing': 3, 'Stucco': 3, 'AsbShng': 2, 'Brk Cmn': 3, 'Stone': 5, 'AsphShn': 1, 'ImStucc': 2, 'CBlock': 2,'Other': 1 })
```

```python
df['Exterior2ndEncoded'].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['MSSubClassNormalized'] = scaler.fit_transform(df[['MSSubClass']])
```

```python
df['PropertyScore'] = (df['MSSubClassNormalized'] * 0.3) + (df['MSZoningEncoded'] * 0.2) + \
                      (df['StreetEncoded'] * 0.1) + (df['LotShapeEncoded'] * 0.1) + \
                      (df['LandContourEncoded'] * 0.1) + (df['LotConfigEncoded'] * 0.1) + \
                      (df['LandSlopeEncoded'] * 0.1) + (df['Exterior2ndEncoded'] * 0.3)
```

```python
df.drop(columns=['MSSubClassNormalized','MSZoningEncoded','StreetEncoded','LotShapeEncoded','LandContourEncoded','LotConfigEncoded','LandSlopeEncoded','Exterior2ndEncoded'],inplace=True)
```

```python
df.drop(columns=['Exterior2nd','MSSubClass','MSZoning','Street','LotShape','LandContour','LotConfig','LandSlope'],inplace=True)
```

```python
df['Functional'] = df['Functional'].replace({
    'Typ': 5,   # Typical functionality (no issues)
    'Min2': 4,  # Minor issues (needs slight adjustments)
    'Min1': 3,  # Minor issues (more noticeable but manageable)
    'Mod': 2,   # Moderate issues (affects usability)
    'Maj1': 1,  # Major issues (significant impact)
    'Maj2': 0,  # Severe issues (requires major repairs)
    'Sev': -1   # Extreme issues (almost unusable)
})
df.drop(columns=['FunctionalEncoded'],inplace=True)
```
